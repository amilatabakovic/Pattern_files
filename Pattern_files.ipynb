## Pattern files for asymmetric gratings

This notebook contains a script used to generate pattern files consisting of periodically repeated FIB features (lines and rectangles). The patterns are designed to be milled from left to right, with a fixed XML structure required by the microscope software.

### Fixed pattern structure

```python
def format_r8(value):
    """Format float as FEI-compatible notation."""
    return "{:.14E}".format(value).replace("E+0", "E+").replace("E-0", "E-")


def ptf_header():
    return """<Content Version="FEI PatternFile 1.2">
  <PatternList>
    <Mode xmlns:dt="urn:schemas-microsoft-com:datatypes" dt:dt="string">Serial</Mode>
    <Beam xmlns:dt="urn:schemas-microsoft-com:datatypes" dt:dt="string">Ion</Beam>
    <GasType xmlns:dt="urn:schemas-microsoft-com:datatypes" dt:dt="string"></GasType>
    <GasSetpoint xmlns:dt="urn:schemas-microsoft-com:datatypes" dt:dt="string"></GasSetpoint>
    <GasNeedlePosition xmlns:dt="urn:schemas-microsoft-com:datatypes" dt:dt="string"></GasNeedlePosition>
    <Defocus xmlns:dt="urn:schemas-microsoft-com:datatypes" dt:dt="r8">0</Defocus>
    <PassCount xmlns:dt="urn:schemas-microsoft-com:datatypes" dt:dt="r8">0</PassCount>
    <Blur xmlns:dt="urn:schemas-microsoft-com:datatypes" dt:dt="r8">0</Blur>
  </PatternList>
"""


def write_ptf_file(ptf_filepath, patterns):
    with open(ptf_filepath, "w") as f:
        f.write(ptf_header())
        for pattern in patterns:
            f.write(pattern)
        f.write("</Content>")


def pattern_rectangle(center_x, width, length, total_time, pass_count, depth, center_y, dwell_time, rotation=0):
    return f"""  <PatternRectangle>
    <Application xmlns:dt="urn:schemas-microsoft-com:datatypes" dt:dt="string">Si</Application>
    <Beam xmlns:dt="urn:schemas-microsoft-com:datatypes" dt:dt="string">Ion</Beam>
    <DwellTime xmlns:dt="urn:schemas-microsoft-com:datatypes" dt:dt="r8">{format_r8(dwell_time)}</DwellTime>
    <CenterX xmlns:dt="urn:schemas-microsoft-com:datatypes" dt:dt="r8">{format_r8(center_x)}</CenterX>
    <CenterY xmlns:dt="urn:schemas-microsoft-com:datatypes" dt:dt="r8">{format_r8(center_y)}</CenterY>
    <Rotation xmlns:dt="urn:schemas-microsoft-com:datatypes" dt:dt="r8">{format_r8(rotation)}</Rotation>
    <GasType xmlns:dt="urn:schemas-microsoft-com:datatypes" dt:dt="string"></GasType>
    <GasSetpoint xmlns:dt="urn:schemas-microsoft-com:datatypes" dt:dt="string"></GasSetpoint>
    <GasNeedlePosition xmlns:dt="urn:schemas-microsoft-com:datatypes" dt:dt="string"></GasNeedlePosition>
    <Defocus xmlns:dt="urn:schemas-microsoft-com:datatypes" dt:dt="r8">0</Defocus>
    <RefreshTime xmlns:dt="urn:schemas-microsoft-com:datatypes" dt:dt="r8">0</RefreshTime>
    <TotalTime xmlns:dt="urn:schemas-microsoft-com:datatypes" dt:dt="r8">{format_r8(total_time)}</TotalTime>
    <PassCount xmlns:dt="urn:schemas-microsoft-com:datatypes" dt:dt="r8">{pass_count}</PassCount>
    <Enable xmlns:dt="urn:schemas-microsoft-com:datatypes" dt:dt="string">True</Enable>
    <ExclusionZone xmlns:dt="urn:schemas-microsoft-com:datatypes" dt:dt="string">False</ExclusionZone>
    <Depth xmlns:dt="urn:schemas-microsoft-com:datatypes" dt:dt="r8">{format_r8(depth)}</Depth>
    <Blur xmlns:dt="urn:schemas-microsoft-com:datatypes" dt:dt="r8">0</Blur>
    <InteractionDiameter xmlns:dt="urn:schemas-microsoft-com:datatypes" dt:dt="r8">0</InteractionDiameter>
    <RelativeInteractionDiameter xmlns:dt="urn:schemas-microsoft-com:datatypes" dt:dt="r8">0</RelativeInteractionDiameter>
    <TotalDiameter xmlns:dt="urn:schemas-microsoft-com:datatypes" dt:dt="r8">0.0000000204</TotalDiameter>
    <VolumePerDose xmlns:dt="urn:schemas-microsoft-com:datatypes" dt:dt="r8">0.00000000027</VolumePerDose>
    <Dose xmlns:dt="urn:schemas-microsoft-com:datatypes" dt:dt="r8">0</Dose>
    <SaturationSputterRate xmlns:dt="urn:schemas-microsoft-com:datatypes" dt:dt="r8">0</SaturationSputterRate>
    <TotalVolumeSputterRate xmlns:dt="urn:schemas-microsoft-com:datatypes" dt:dt="r8">0</TotalVolumeSputterRate>
    <SaturationCurrentDensity xmlns:dt="urn:schemas-microsoft-com:datatypes" dt:dt="r8">1</SaturationCurrentDensity>
    <MaximumDosePerArea xmlns:dt="urn:schemas-microsoft-com:datatypes" dt:dt="r8">0</MaximumDosePerArea>
    <VolumeEtchRate xmlns:dt="urn:schemas-microsoft-com:datatypes" dt:dt="r8">0</VolumeEtchRate>
    <VolumeGasAssistedRate xmlns:dt="urn:schemas-microsoft-com:datatypes" dt:dt="r8">0</VolumeGasAssistedRate>
    <VolumeDoseLimitedRate xmlns:dt="urn:schemas-microsoft-com:datatypes" dt:dt="r8">0</VolumeDoseLimitedRate>
    <TotalLinearSputterRate xmlns:dt="urn:schemas-microsoft-com:datatypes" dt:dt="r8">0</TotalLinearSputterRate>
    <TotalBeamArea xmlns:dt="urn:schemas-microsoft-com:datatypes" dt:dt="r8">0</TotalBeamArea>
    <DepthPerPass xmlns:dt="urn:schemas-microsoft-com:datatypes" dt:dt="r8">0</DepthPerPass>
    <AreaCalculation xmlns:dt="urn:schemas-microsoft-com:datatypes" dt:dt="string">Pattern</AreaCalculation>
    <Width xmlns:dt="urn:schemas-microsoft-com:datatypes" dt:dt="r8">{format_r8(width)}</Width>
    <Length xmlns:dt="urn:schemas-microsoft-com:datatypes" dt:dt="r8">{format_r8(length)}</Length>
    <ScanDirection xmlns:dt="urn:schemas-microsoft-com:datatypes" dt:dt="string">DynamicLeftToRight</ScanDirection>
    <ScanType xmlns:dt="urn:schemas-microsoft-com:datatypes" dt:dt="string">Serpentine</ScanType>
    <PitchX xmlns:dt="urn:schemas-microsoft-com:datatypes" dt:dt="r8">0.0000000153</PitchX>
    <PitchY xmlns:dt="urn:schemas-microsoft-com:datatypes" dt:dt="r8">0.0000000153</PitchY>
    <EqualPitch xmlns:dt="urn:schemas-microsoft-com:datatypes" dt:dt="string">True</EqualPitch>
    <OverlapX xmlns:dt="urn:schemas-microsoft-com:datatypes" dt:dt="r8">25</OverlapX>
    <OverlapY xmlns:dt="urn:schemas-microsoft-com:datatypes" dt:dt="r8">25</OverlapY>
    <SelectiveMilling>
      <IsSelectiveMillingEnabled xmlns:dt="urn:schemas-microsoft-com:datatypes" dt:dt="string">False</IsSelectiveMillingEnabled>
      <TimeIntervalBeforeAdjust xmlns:dt="urn:schemas-microsoft-com:datatypes" dt:dt="r8">2</TimeIntervalBeforeAdjust>
      <ContrastThresholdMinimum xmlns:dt="urn:schemas-microsoft-com:datatypes" dt:dt="r8">0</ContrastThresholdMinimum>
      <ContrastThresholdMaximum xmlns:dt="urn:schemas-microsoft-com:datatypes" dt:dt="r8">1</ContrastThresholdMaximum>
    </SelectiveMilling>
    <EndpointDetection>
      <IsEndpointDetectionEnabled xmlns:dt="urn:schemas-microsoft-com:datatypes" dt:dt="string">False</IsEndpointDetectionEnabled>
      <EndpointDetectionSignalType xmlns:dt="urn:schemas-microsoft-com:datatypes" dt:dt="string">Intensity</EndpointDetectionSignalType>
      <EndpointDetectionEndpointType xmlns:dt="urn:schemas-microsoft-com:datatypes" dt:dt="string">Peak</EndpointDetectionEndpointType>
      <EndpointNum xmlns:dt="urn:schemas-microsoft-com:datatypes" dt:dt="r8">0</EndpointNum>
    </EndpointDetection>
  </PatternRectangle>
"""

def pattern_line(center_x, length, total_time, pass_count, depth, center_y, dwell_time, rotation=4.71238898038469):
    return f"""  <PatternLine>
    <Application xmlns:dt="urn:schemas-microsoft-com:datatypes" dt:dt="string">Si</Application>
    <Beam xmlns:dt="urn:schemas-microsoft-com:datatypes" dt:dt="string">Ion</Beam>
    <DwellTime xmlns:dt="urn:schemas-microsoft-com:datatypes" dt:dt="r8">{format_r8(dwell_time)}</DwellTime>
    <CenterX xmlns:dt="urn:schemas-microsoft-com:datatypes" dt:dt="r8">{format_r8(center_x)}</CenterX>
    <CenterY xmlns:dt="urn:schemas-microsoft-com:datatypes" dt:dt="r8">{format_r8(center_y)}</CenterY>
    <Rotation xmlns:dt="urn:schemas-microsoft-com:datatypes" dt:dt="r8">{format_r8(rotation)}</Rotation>
    <GasType xmlns:dt="urn:schemas-microsoft-com:datatypes" dt:dt="string"></GasType>
    <Defocus xmlns:dt="urn:schemas-microsoft-com:datatypes" dt:dt="r8">0</Defocus>
    <RefreshTime xmlns:dt="urn:schemas-microsoft-com:datatypes" dt:dt="r8">0</RefreshTime>
    <TotalTime xmlns:dt="urn:schemas-microsoft-com:datatypes" dt:dt="r8">{format_r8(total_time)}</TotalTime>
    <PassCount xmlns:dt="urn:schemas-microsoft-com:datatypes" dt:dt="r8">{pass_count}</PassCount>
    <Enable xmlns:dt="urn:schemas-microsoft-com:datatypes" dt:dt="string">True</Enable>
    <ExclusionZone xmlns:dt="urn:schemas-microsoft-com:datatypes" dt:dt="string">False</ExclusionZone>
    <Depth xmlns:dt="urn:schemas-microsoft-com:datatypes" dt:dt="r8">{format_r8(depth)}</Depth>
    <Blur xmlns:dt="urn:schemas-microsoft-com:datatypes" dt:dt="r8">0</Blur>
    <InteractionDiameter xmlns:dt="urn:schemas-microsoft-com:datatypes" dt:dt="r8">0</InteractionDiameter>
    <RelativeInteractionDiameter xmlns:dt="urn:schemas-microsoft-com:datatypes" dt:dt="r8">0</RelativeInteractionDiameter>
    <TotalDiameter xmlns:dt="urn:schemas-microsoft-com:datatypes" dt:dt="r8">0.000000031</TotalDiameter>
    <VolumePerDose xmlns:dt="urn:schemas-microsoft-com:datatypes" dt:dt="r8">0.00000000015</VolumePerDose>
    <Dose xmlns:dt="urn:schemas-microsoft-com:datatypes" dt:dt="r8">0</Dose>
    <SaturationSputterRate xmlns:dt="urn:schemas-microsoft-com:datatypes" dt:dt="r8">0</SaturationSputterRate>
    <TotalVolumeSputterRate xmlns:dt="urn:schemas-microsoft-com:datatypes" dt:dt="r8">0</TotalVolumeSputterRate>
    <SaturationCurrentDensity xmlns:dt="urn:schemas-microsoft-com:datatypes" dt:dt="r8">1</SaturationCurrentDensity>
    <MaximumDosePerArea xmlns:dt="urn:schemas-microsoft-com:datatypes" dt:dt="r8">0</MaximumDosePerArea>
    <VolumeEtchRate xmlns:dt="urn:schemas-microsoft-com:datatypes" dt:dt="r8">0</VolumeEtchRate>
    <VolumeGasAssistedRate xmlns:dt="urn:schemas-microsoft-com:datatypes" dt:dt="r8">0</VolumeGasAssistedRate>
    <VolumeDoseLimitedRate xmlns:dt="urn:schemas-microsoft-com:datatypes" dt:dt="r8">0</VolumeDoseLimitedRate>
    <TotalLinearSputterRate xmlns:dt="urn:schemas-microsoft-com:datatypes" dt:dt="r8">0</TotalLinearSputterRate>
    <TotalBeamArea xmlns:dt="urn:schemas-microsoft-com:datatypes" dt:dt="r8">0</TotalBeamArea>
    <DepthPerPass xmlns:dt="urn:schemas-microsoft-com:datatypes" dt:dt="r8">0</DepthPerPass>
    <Length xmlns:dt="urn:schemas-microsoft-com:datatypes" dt:dt="r8">{format_r8(length)}</Length>
    <Pitch xmlns:dt="urn:schemas-microsoft-com:datatypes" dt:dt="r8">0.0000000155</Pitch>
    <Overlap xmlns:dt="urn:schemas-microsoft-com:datatypes" dt:dt="r8">50</Overlap>
    <SelectiveMilling>
      <IsSelectiveMillingEnabled xmlns:dt="urn:schemas-microsoft-com:datatypes" dt:dt="string">False</IsSelectiveMillingEnabled>
      <TimeIntervalBeforeAdjust xmlns:dt="urn:schemas-microsoft-com:datatypes" dt:dt="r8">2</TimeIntervalBeforeAdjust>
      <ContrastThresholdMinimum xmlns:dt="urn:schemas-microsoft-com:datatypes" dt:dt="r8">0</ContrastThresholdMinimum>
      <ContrastThresholdMaximum xmlns:dt="urn:schemas-microsoft-com:datatypes" dt:dt="r8">1</ContrastThresholdMaximum>
    </SelectiveMilling>
    <EndpointDetection>
      <IsEndpointDetectionEnabled xmlns:dt="urn:schemas-microsoft-com:datatypes" dt:dt="string">False</IsEndpointDetectionEnabled>
      <EndpointDetectionSignalType xmlns:dt="urn:schemas-microsoft-com:datatypes" dt:dt="string">Intensity</EndpointDetectionSignalType>
      <EndpointDetectionEndpointType xmlns:dt="urn:schemas-microsoft-com:datatypes" dt:dt="string">Peak</EndpointDetectionEndpointType>
      <EndpointNum xmlns:dt="urn:schemas-microsoft-com:datatypes" dt:dt="r8">0</EndpointNum>
    </EndpointDetection>
  </PatternLine>
"""
```

### Pattern Files Consisting of Two Rectangles

```python
def generate_two_rectangles(
    ptf_filepath,
    period_pitch,
    spacing,
    pattern_length,
    center_y,
    dwell_time,
    rect1_width,
    rect1_length,
    rect1_total_time,
    rect1_pass_count,
    rect1_depth,
    rect2_width,
    rect2_length,
    rect2_total_time,
    rect2_pass_count,
    rect2_depth,
):
    """Generate a PTF file with two rectangles per unit cell.

    Args:
        ptf_filepath: Output path for the PTF file.
        period_pitch: Distance between repeated unit cells along x.
        spacing: Offset between the first and second rectangle within each unit cell.
        pattern_length: Total length of the patterned region along x.
        center_y: Fixed y position for all features.
        dwell_time: Beam dwell time used for each feature.
        rect1_width: Width of the first rectangle.
        rect1_length: Length of the first rectangle.
        rect1_total_time: Total beam exposure time for the first rectangle.
        rect1_pass_count: Number of passes for the first rectangle.
        rect1_depth: Effective milling depth for the first rectangle.
        rect2_width: Width of the second rectangle.
        rect2_length: Length of the second rectangle.
        rect2_total_time: Total beam exposure time for the second rectangle.
        rect2_pass_count: Number of passes for the second rectangle.
        rect2_depth: Effective milling depth for the second rectangle.
    """
    num_periods = round(pattern_length / period_pitch)
    patterns = []

    for n in range(num_periods):
        base_x = n * period_pitch

        patterns.append(pattern_rectangle(
            base_x,
            rect1_width,
            rect1_length,
            rect1_total_time,
            rect1_pass_count,
            rect1_depth,
            center_y,
            dwell_time,
        ))

        patterns.append(pattern_rectangle(
            base_x + spacing,
            rect2_width,
            rect2_length,
            rect2_total_time,
            rect2_pass_count,
            rect2_depth,
            center_y,
            dwell_time,
        ))

    write_ptf_file(ptf_filepath, patterns)
```

### Pattern Files Consisting of One Line and One Rectangle 

```python
def generate_line_rectangle(
    ptf_filepath,
    period_pitch,
    spacing,
    pattern_length,
    center_y,
    dwell_time,
    line_length,
    line_total_time,
    line_pass_count,
    line_depth,
    rect2_width,
    rect2_length,
    rect2_total_time,
    rect2_pass_count,
    rect2_depth,
    start_x=0.0,
):
    """Generate a PTF file with one line and one rectangle per unit cell.

    Args:
        ptf_filepath: Output path for the PTF file.
        period_pitch: Distance between repeated unit cells along x.
        spacing: Offset between the line and rectangle within each unit cell.
        pattern_length: Total length of the patterned region along x.
        center_y: Fixed y position for all features.
        dwell_time: Beam dwell time used for each feature.
        line_length: Length of the line feature.
        line_total_time: Total beam exposure time for the line feature.
        line_pass_count: Number of passes for the line feature.
        line_depth: Effective milling depth for the line feature.
        rect2_width: Width of the rectangle feature.
        rect2_length: Length of the rectangle feature.
        rect2_total_time: Total beam exposure time for the rectangle feature.
        rect2_pass_count: Number of passes for the rectangle feature.
        rect2_depth: Effective milling depth for the rectangle feature.
        start_x: Starting x-coordinate for the patterned region.
    """
    num_periods = round(pattern_length / period_pitch)
    patterns = []

    for n in range(num_periods):
        base_x = start_x + n * period_pitch

        patterns.append(pattern_line(
            base_x,
            line_length,
            line_total_time,
            line_pass_count,
            line_depth,
            center_y,
            dwell_time,
        ))

        patterns.append(pattern_rectangle(
            base_x + spacing,
            rect2_width,
            rect2_length,
            rect2_total_time,
            rect2_pass_count,
            rect2_depth,
            center_y,
            dwell_time,
        ))

    write_ptf_file(ptf_filepath, patterns)
```

### Pattern Files Consisting of Two Lines

```python
def generate_two_lines(
    ptf_filepath,
    period_pitch,
    spacing,
    pattern_length,
    center_y,
    dwell_time,
    line_length,
    line_total_time,
    line_pass_count,
    line_depth,
    start_x=0.0,
):
    """Generate a PTF file with two lines per unit cell.

    Args:
        ptf_filepath: Output path for the PTF file.
        period_pitch: Distance between repeated unit cells along x.
        spacing: Offset between the two lines in each unit cell.
        pattern_length: Total length of the patterned region along x.
        center_y: Fixed y position for all features.
        dwell_time: Beam dwell time used for each feature.
        line_length: Length of the line features.
        line_total_time: Total beam exposure time for the line features.
        line_pass_count: Number of passes for the line features.
        line_depth: Effective milling depth for the line features.
        start_x: Starting x-coordinate for the patterned region.
    """
    num_periods = round(pattern_length / period_pitch)
    patterns = []

    for n in range(num_periods):
        base_x = start_x + n * period_pitch

        patterns.append(pattern_line(
            base_x,
            line_length,
            line_total_time,
            line_pass_count,
            line_depth,
            center_y,
            dwell_time,
        ))

        patterns.append(pattern_line(
            base_x + spacing,
            line_length,
            line_total_time,
            line_pass_count,
            line_depth,
            center_y,
            dwell_time,
        ))

    write_ptf_file(ptf_filepath, patterns)
```
